## Now, Extracting totals from Time Series Data

```python
df2 = pd.read_json(r"C:\Users\Piyush\OneDrive\Desktop\Data Science\1 Masai Evaluation and assignment data\Masai projects\COVID 19\data2.json")
```

```python
final_df = pd.DataFrame()
```

```python
df2
```

```python
ind = df2.loc['dates'].index
for sn in ind:
  if sn == 'UN': # Error State Name, skipping it
    continue
  else:
    demo = pd.DataFrame(df2.loc['dates'][sn])
    dates = pd.DataFrame(demo.loc['total']).index
    new_data = []
    for i in pd.DataFrame(demo.loc['total'])['total']: # Forming dataframe from the data
      temp = i
      temp['state'] = sn                               # Creating a state column with the state name
      new_data.append(temp)
    temp_df = pd.DataFrame(new_data, index = dates)    # Forming the dataframe for the selected state
    final_df = final_df.append(temp_df)                # Appending this new dataframe to the original dataframe
```
